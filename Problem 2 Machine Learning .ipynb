## 2.2

```python
import nltk
from nltk.corpus import stopwords 
```

```python
stop_words = stopwords.words('english')
add_to_stop_words = [',','.','-',';',':','--',"'",'(',')']
stop_words.extend(add_to_stop_words)
stop_words = set(stop_words)
```

```python
stop_words
```

```python
words1 = list(map(lambda x:x.lower(), (inaugural.words(fileids = '1941-Roosevelt.txt'))))
words2 = list(map(lambda x:x.lower(), (inaugural.words(fileids = '1961-Kennedy.txt'))))
words3 = list(map(lambda x:x.lower(), (inaugural.words(fileids = '1973-Nixon.txt'))))
```

```python
nltk.FreqDist(words1)
```

```python
nltk.FreqDist(words2)
```

```python
nltk.FreqDist(words3)
```

## Removing stopwords from Roosevelt's speech

```python
for i in stop_words:
    if i in words1:
        while i in words1:
            words1.remove(i)
```

```python
dict_words1 = nltk.FreqDist(words1)
dict_words1
```

## Removing stopwords from Kennedy's speech

```python
for i in stop_words:
    if i in words2:
        while i in words2:
            words2.remove(i)
```

```python
dict_words2 = nltk.FreqDist(words2)
dict_words2
```

## Removing stopwords from Nixon's speech

```python
for i in stop_words:
    if i in words3:
        while i in words3:
            words3.remove(i)
```

```python
dict_words3 = nltk.FreqDist(words3)
dict_words3
```

## 2.3

```python
from collections import Counter
```

```python
k1 = Counter(dict_words1)
high1 = k1.most_common(3)
```

```python
print("Most frequent words in Roosevelt's speech:") 
for i in high1: 
    print(i[0]," :",i[1]," ")
```

```python
k2 = Counter(dict_words2)
high2 = k2.most_common(3)
```

```python
print("Most frequent words in Kennedy's speech:") 
for i in high2: 
    print(i[0]," :",i[1]," ")
```

```python
k3 = Counter(dict_words3)
high3 = k3.most_common(3)
```

```python
print("Most frequent words in Nixon's speech:") 
for i in high3: 
    print(i[0]," :",i[1]," ")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt   
import matplotlib.style
```

## Wordcloud of Roosevelt's speech

```python
text1 = inaugural.raw('1941-Roosevelt.txt')
wordcloud1 = WordCloud(width = 1000, height = 1000, background_color ='white',max_font_size=250).generate(text1)
plt.figure(figsize=(10, 16), facecolor = None)
# plot wordcloud in matplotlib
plt.imshow(wordcloud1, interpolation="bilinear")
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0)
plt.show()
```

## Wordcloud of Kennedy's speech

```python
text2 = inaugural.raw('1961-Kennedy.txt')
wordcloud2 = WordCloud(width = 1000, height = 1000, background_color ='white',max_font_size=250).generate(text2)
plt.figure(figsize=(10, 16), facecolor = None)
# plot wordcloud in matplotlib
plt.imshow(wordcloud2, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
